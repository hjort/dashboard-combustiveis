```python
import pandas as pd
```

```python
# ler os dados do arquivo CSV
df = pd.read_csv("ge-2017-2022.csv.bz2",
                 dtype={'regiao': 'category', 'uf': 'category', 
                        'municipio': 'category', 'bairro': 'category',
                        'produto': 'category', 'venda': 'float32',
                        'compra': 'float32', 'unidade': 'category',
                        'ano': 'uint16', 'mes': 'uint8',
                        'bandeira': 'category'})
```

```python
df
```

```python
# selecionar apenas linhas que possuem valor de compra preenchido
df2 = df[~df['compra'].isna()]
df2
```

```python
# remover colunas que serÃ£o desprezadas
df2r = df2.drop(['regiao', 'municipio', 'bairro', 'bandeira', 'revenda'], axis=1)
#df2r = df2.drop(['uf', 'municipio', 'bairro', 'bandeira', 'revenda'], axis=1)
df2r
```

```python
df2r.describe(include='all')
```

```python
df2r.corr()
```

```python
# converter colunas categÃ³ricas em numÃ©ricas (one-hot encoding)
df2r = pd.get_dummies(df2r)
df2r
```

```python
df2r.info()
```

```python
df2r.corr()
```

## Treinar modelo de regressÃ£o

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
X = df2r.drop('compra', axis=1).values
y = df2r['compra']
```

```python
X.shape, y.shape
```

```python
modelo = RandomForestRegressor(random_state=42, n_jobs=-1)
modelo
```

```python
%time modelo.fit(X, y)
```

```python
X[0]
```

```python
modelo.predict([X[0]])
```

## Imputar valores ausentes

```python
# selecionar apenas linhas que nÃ£o possuem valor de compra preenchido
df3 = df[df['compra'].isna()]
df3
```

```python
# remover colunas que serÃ£o desprezadas
df3r = df3.drop(['regiao', 'municipio', 'bairro', 'bandeira', 'revenda', 'compra'], axis=1)
#df3r = df3.drop(['uf', 'municipio', 'bairro', 'bandeira', 'revenda', 'compra'], axis=1)
df3r
```

```python
# converter colunas categÃ³ricas em numÃ©ricas (one-hot encoding)
df3r = pd.get_dummies(df3r)
df3r
```

```python
df3r.info()
```

```python
df3r.corr()
```

```python
X3 = df3r.values
X3
```

```python
y3_pred = modelo.predict(X3)
y3_pred
```

```python
# imputar os valores ausentes
df3['compra'] = y3_pred
df3
```

```python
df3.corr()
```

## Gerar dataframe consolidado

```python
# fazer marcaÃ§Ã£o de que certos valores foram imputados
df = df2.copy()
df['imputado'] = 0
df = pd.concat([df, df3])
df['imputado'] = df['imputado'].fillna(1)
df['imputado'] = df['imputado'].astype('uint8')
df
```

```python
#df = pd.concat([df2, df3])
#df
```

```python
# arredondar valores para 4 casas decimais
for col in ['venda', 'compra']:
    df[col] = df[col].round(3)
df
```

```python
# gerar arquivo CSV com valores imputados
arquivo_saida = "ge-cv-2017-2022.csv.bz2"
df.to_csv(arquivo_saida, index=False, encoding='utf-8')
```
