```python
# ler os dados do arquivo CSV
df = pd.read_csv("ca-2015-2022.csv.bz2",
                 dtype={'regiao': 'category', 'uf': 'category', 
                        'municipio': 'category', 'bairro': 'category',
                        'produto': 'category', 'venda': 'float32',
                        'compra': 'float32', 'unidade': 'category',
                        'ano': 'category', 'mes': 'category', 'dia': 'category'})
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
for col in df.columns:
    print(df[[col]].value_counts(), "\n")
```

```python
total = len(df)
for col in df.columns:
    qtd = df[df[col].isna()]['uf'].count()
    print("%10s: %7d nulos (%6.2f%%)" % (col, qtd, qtd / total * 100))
print("%10s: %7d linhas" % ("total", total))
```

```python
df.groupby(['produto', 'unidade'])[['venda']].count().T
```

```python
df[~df['compra'].isna()].\
    groupby(['ano'])[['compra']].count().T
```

```python
df[df['produto'].isin(['GASOLINA', 'ETANOL'])].\
    groupby(['ano'])[['compra']].count().T
```
