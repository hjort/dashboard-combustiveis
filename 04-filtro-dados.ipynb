```python
# ler os dados do arquivo CSV
df = pd.read_csv("ca-2017-2022.csv.bz2",
                 dtype={'regiao': 'category', 'uf': 'category', 
                        'municipio': 'category', 'bairro': 'category',
                        'produto': 'category', 'venda': 'float32',
                        'compra': 'float32', 'unidade': 'category',
                        'ano': 'category', 'mes': 'category', 'dia': 'category'})
```

```python
# manter apenas 2 tipos de produtos e remover unidade de medida
df = df[df['produto'].isin(['GASOLINA', 'ETANOL'])].drop('unidade', axis=1)
df
```

```python
# remover campo dia (irrelevante para a anÃ¡lise)
df.drop('dia', axis=1, inplace=True)
df
```
